```python
from pypathway import *
```

```python
path = PublicDatabase.search_wp("jak")[0].load()
```

```python
# screen pathway info
path
```

```python
# detail information
print(path.summary())
```

```python
# node operation
path.children
```

```python
# father: return super node of a certain node
path.children[0].father == path
```

```python
# root: return the root node of certain node
path.children[1].children[0].root == path
```

```python
# element related method
path.get_element_by_class("DataNode")
```

```python
# get element by GraphId, return a list
path.get_element_by_id("bf41f")
```

```python
# get element by name, the TextLabel in WikiPathway pathway
path.get_element_by_label("IRS1")
```

```python
# if there are no spceial characters in label, you can simplily using .label to get the node
path.IRS1
```

```python
# get element by type, The Type prop in WikiPathway pathway
path.get_element_by_type("GeneProduct")
```

```python
# get element by ontology id
path.get_element_by_oid("6776")
```

```python
# quick method: memebrs, return all node
path.members
```

```python
# quick method: entities (aka. nodes), return all DataNodes
path.entities
```

```python
# qucik method: reaction, (aka, class == interaction)
path.reactions
```
